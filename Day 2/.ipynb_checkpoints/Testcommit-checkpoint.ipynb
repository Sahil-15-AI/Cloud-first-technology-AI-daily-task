### even or odd 

```python
def even_odd(num):
    if num % 2 == 0:
        return "The given number is even"
    else:
        return "The given number is odd"

number = int(input("Enter a number"))
even_odd((number))
```

### reverse the string 

```python
def reverse_string(var):
    return var[:: -1]

var_string = input(" enter a string ")

reverse_string((var_string))
```

## Palindrome 


```python
def palindrome(var):
    reverse = var[:: -1]

    if reverse == var:
        return "The given String is palindrome"
    else:
        return "The given String is not a palindrome"
        

string = input(" Enter a string ")

palindrome((string))
```
