## 1. Classes, Objects & Constructors
A **class** is a blueprint; an **object** is an instance of it. `__init__` is the constructor — it runs automatically when a new object is created.

````python
class Book:
    def __init__(self, title, author, price):
        self.title = title
        self.author = author
        self.price = price

    def __str__(self):
        return f"{self.title} by {self.author} (${self.price})"

b = Book("Dune", "Frank Herbert", 25)
print(b)
````

## 2. Inheritance & Polymorphism
**Inheritance** lets a class reuse and extend another class's behavior. **Polymorphism** means the same method name behaves differently depending on the object's actual class.

````python
class Shape:
    def area(self):
        raise NotImplementedError

class Circle(Shape):
    def __init__(self, r):
        self.r = r
    def area(self):
        return 3.14159 * self.r ** 2

class Square(Shape):
    def __init__(self, side):
        self.side = side
    def area(self):
        return self.side ** 2

shapes = [Circle(3), Square(4)]
for s in shapes:
    print(type(s).__name__, "->", s.area())   # same method call, different behavior
````

## 3. Encapsulation & Abstraction
**Encapsulation**: hide internal state behind `_private` attributes and expose controlled access via `@property`.

**Abstraction**: use `ABC` + `@abstractmethod` to define a contract that subclasses must fulfil, without instantiating the base class itself.

````python
from abc import ABC, abstractmethod

class Account:
    def __init__(self, balance):
        self._balance = balance   # "protected" by convention

    @property
    def balance(self):
        return self._balance

acc = Account(1000)
print(acc.balance)     # read access via property
# acc.balance = 5000   # would fail: no setter defined -> encapsulation enforced
````

````python
class PaymentMethod(ABC):
    @abstractmethod
    def pay(self, amount):
        ...

class UPI(PaymentMethod):
    def pay(self, amount):
        return f"Paid {amount} via UPI"

print(UPI().pay(500))
try:
    PaymentMethod()   # cannot instantiate an abstract class
except TypeError as e:
    print("Error:", e)
````

## 4. Iterators & Generators
A custom **iterator** implements `__iter__` and `__next__`. A **generator** is a simpler way to get the same result using `yield`, without building the whole sequence in memory.

````python
class Countdown:
    def __init__(self, start):
        self.current = start
    def __iter__(self):
        return self
    def __next__(self):
        if self.current <= 0:
            raise StopIteration
        val = self.current
        self.current -= 1
        return val

print(list(Countdown(5)))
````

````python
def fibonacci(n):
    a, b = 0, 1
    for _ in range(n):
        yield a
        a, b = b, a + b

print(list(fibonacci(10)))
````

## 5. Decorators
A decorator wraps a function to add behavior (logging, timing, validation, retries) without changing the function's own code.

````python
import functools, time

def timer(func):
    @functools.wraps(func)
    def wrapper(*args, **kwargs):
        start = time.perf_counter()
        result = func(*args, **kwargs)
        print(f"{func.__name__} took {time.perf_counter()-start:.6f}s")
        return result
    return wrapper

@timer
def slow_square(n):
    return n * n

print(slow_square(12))
````

## 6. Context Managers
Anything used in a `with` block needs `__enter__`/`__exit__`. This guarantees setup/teardown (e.g. closing a file, saving data) happens even if an exception is raised inside the block.

````python
class Timer:
    def __enter__(self):
        self.start = time.perf_counter()
        return self
    def __exit__(self, exc_type, exc_val, exc_tb):
        print(f"Block took {time.perf_counter()-self.start:.6f}s")
        return False

with Timer():
    total = sum(range(1_000_000))
print(total)
````

## 7. Logging
Prefer `logging` over `print()` for real applications: it supports severity levels, timestamps, and can be routed to files or external systems.

````python
import logging
logging.basicConfig(level=logging.INFO, format="%(asctime)s %(levelname)s: %(message)s")
log = logging.getLogger("demo")

log.debug("This won't show (below INFO level)")
log.info("Application started")
log.warning("Something looks off, but we can continue")
log.error("Something failed")
````

## 8. JSON Handling
`json.dumps`/`json.loads` convert between Python objects and JSON strings; `json.dump`/`json.load` do the same directly with files.

````python
import json

data = {"name": "Aditi", "role": "Developer", "skills": ["Python", "SQL"]}

json_str = json.dumps(data, indent=2)
print(json_str)

parsed = json.loads(json_str)
print(parsed["skills"])
````

## 9. REST API Consumption using `requests`
`requests.get()` fetches data from a REST API; `.raise_for_status()` turns HTTP errors into Python exceptions, and `.json()` parses the response body.

````python
import requests

try:
    response = requests.get("https://api.github.com/users/octocat", timeout=10)
    response.raise_for_status()
    data = response.json()
    print(data["login"], "-", data["public_repos"], "public repos")
except requests.exceptions.RequestException as e:
    # GitHub's public API is rate-limited per IP; this keeps the demo robust
    print("Request failed (likely rate-limited):", e)
````
