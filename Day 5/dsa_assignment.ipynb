## 1. Time & Space Complexity — Big-O Notation

**Big-O notation** describes how the runtime (or memory usage) of an algorithm grows as the input size `n` grows. It describes the *worst-case* growth rate, ignoring constants.

Common complexities, from fastest to slowest growth:

| Big-O | Name | Example |
|---|---|---|
| O(1) | Constant | Accessing `arr[0]` |
| O(log n) | Logarithmic | Binary search |
| O(n) | Linear | Looping through a list once |
| O(n log n) | Linearithmic | Merge sort, Quick sort (avg) |
| O(n²) | Quadratic | Nested loops (Bubble sort) |
| O(2^n) | Exponential | Naive recursive Fibonacci |

**Space complexity** is the same idea, but for memory used (extra variables, recursion stack, data structures created), not runtime.

### Problem 2.1 — Reverse an array **in place**
Write `reverse_array(arr)` that reverses the list **without using `arr[::-1]` or `.reverse()`** — use a two-pointer approach instead. Modify `arr` in place and return it.

```python
def reverse_array(arr):
    # TODO: use two pointers (start, end) swapping until they meet
    start = 0
    end = len(arr)-1
    while start<end:
        arr[start], arr[end] = arr[end], arr[start]
        start += 1
        end -= 1
    return arr
    pass
```

### Problem 2.3 — Find the maximum subarray sum (Kadane's Algorithm)
Given an array of integers (possibly negative), find the contiguous subarray with the largest sum. Return the sum only.

Example: `[-2,1,-3,4,-1,2,1,-5,4]` → `6` (subarray `[4,-1,2,1]`)

```python
def max_subarray_sum(arr):
    # TODO: implement Kadane's algorithm in O(n) time
    current_sum = arr[0]
    max_sum = arr[0]

    for i in range(1, len(arr)):
        current_sum = max(arr[i] , current_sum + arr[i])
        max_sum = max(max_sum, current_sum)
    return max_sum
    pass
```

### Problem 3.3 — Reverse a Linked List
Write `reverse_linked_list(ll)` which reverses a `LinkedList` **in place** (re-point the `.next` pointers, don't just build a new list from `to_list()`) and returns the same `ll` object.

```python
def reverse_linked_list(ll):
    prev = None
    current = ll.head
    while current:
        next_node = current.next
        current.next = prev
        prev = current
        current = next_node
    ll.head = prev
    return ll
```

## 4. Stacks

A **stack** is a LIFO (Last In, First Out) structure. Core operations: `push`, `pop`, `peek`, `is_empty` — all O(1).

### Problem 4.1 — Implement a Stack
Implement it using a Python list internally.

```python
class Stack:
    def __init__(self):
        self.items = []

    def push(self, value):
        self.items.append(value)

    def pop(self):
        if self.is_empty():
            return None
        return self.items.pop()

    def peek(self):
        if self.is_empty():
            return None
        return self.items[-1]

    def is_empty(self):
        return len(self.items) == 0
```

```python
# Test cell
s = Stack()
assert s.is_empty() == True
s.push(1)
s.push(2)
s.push(3)
assert s.peek() == 3
assert s.pop() == 3
assert s.pop() == 2
assert s.is_empty() == False
assert s.pop() == 1
assert s.is_empty() == True
assert s.pop() is None
print("All tests passed!")
```

### Problem 4.2 — Valid Parentheses
Given a string containing only `(`, `)`, `{`, `}`, `[`, `]`, determine if the brackets are balanced and correctly nested. Use your `Stack` class from above.

Example: `"({[]})"` → `True`, `"(]"` → `False`

```python
def is_valid_parentheses(s):
    stack = Stack()
    matching = {')': '(', '}': '{', ']': '['}
    for ch in s:
        if ch in '({[':
            stack.push(ch)
        elif ch in ')}]':
            if stack.is_empty() or stack.pop() != matching[ch]:
                return False
    return stack.is_empty()
```

## 5. Queues

A **queue** is a FIFO (First In, First Out) structure. Core operations: `enqueue`, `dequeue`, `peek`, `is_empty`.

### Problem 5.1 — Implement a Queue
Use `collections.deque` internally for O(1) enqueue/dequeue (avoid using a plain list with `.pop(0)`, which is O(n)).

```python
from collections import deque

class Queue:
    def __init__(self):
        self.items = deque()

    def enqueue(self, value):
        self.items.append(value)

    def dequeue(self):
        if self.is_empty():
            return None
        return self.items.popleft()

    def peek(self):
        if self.is_empty():
            return None
        return self.items[0]

    def is_empty(self):
        return len(self.items) == 0
```

```python
# Test cell
q = Queue()
assert q.is_empty() == True
q.enqueue(1)
q.enqueue(2)
q.enqueue(3)
assert q.peek() == 1
assert q.dequeue() == 1
assert q.dequeue() == 2
assert q.is_empty() == False
assert q.dequeue() == 3
assert q.is_empty() == True
print("All tests passed!")
```

### Problem 5.2 — Implement a Queue using Two Stacks
A classic interview question: implement a Queue's `enqueue`/`dequeue` behaviour using **two stacks** instead of a deque. This tests your understanding of both structures.

Hint: use `stack_in` for enqueues, and move everything to `stack_out` when you need to dequeue (only if `stack_out` is empty).

```python
class QueueWithStacks:
    def __init__(self):
        self.stack_in = []
        self.stack_out = []

    def enqueue(self, value):
        self.stack_in.append(value)

    def dequeue(self):
        if not self.stack_out:
            while self.stack_in:
                self.stack_out.append(self.stack_in.pop())
        if not self.stack_out:
            return None
        return self.stack_out.pop()
```

## 6. Hash Maps / Dictionaries

A **hash map** gives average **O(1)** lookup, insert, and delete by mapping keys to values via a hash function. Python's `dict` is a hash map.

### Problem 6.1 — Two Sum
Given an array `nums` and a target integer, return the indices of the two numbers that add up to `target`. Assume exactly one solution exists. Solve it in **O(n)** time using a dictionary (not a nested loop).

Example: `nums=[2,7,11,15], target=9` → `[0,1]`

```python
def two_sum(nums, target):
    seen = {}
    for i, num in enumerate(nums):
        complement = target - num
        if complement in seen:
            return [seen[complement], i]
        seen[num] = i
```

```python
# Test cell
assert sorted(two_sum([2,7,11,15], 9)) == [0,1]
assert sorted(two_sum([3,2,4], 6)) == [1,2]
assert sorted(two_sum([3,3], 6)) == [0,1]
print("All tests passed!")
```

### Problem 6.2 — First Non-Repeating Character
Given a string, return the index of the first character that does not repeat. Return `-1` if every character repeats.

Example: `"swiss"` → `1` (the `'w'`)

```python
def first_unique_char(s):
    freq = {}
    for ch in s:
        freq[ch] = freq.get(ch, 0) + 1
    for i, ch in enumerate(s):
        if freq[ch] == 1:
            return i
    return -1
```

### Problem 6.3 — Group Anagrams
Given a list of strings, group the anagrams together. Return a list of lists (order of groups/elements doesn't matter, tests will sort before comparing).

Example: `["eat","tea","tan","ate","nat","bat"]` → `[["eat","tea","ate"],["tan","nat"],["bat"]]`

```python
def group_anagrams(strs):
    groups = {}
    for word in strs:
        key = ''.join(sorted(word))
        groups.setdefault(key, []).append(word)
    return list(groups.values())
```

## 7. Trees (Binary Tree, BST)

A **binary tree** is a hierarchical structure where each node has at most 2 children (`left`, `right`).

A **Binary Search Tree (BST)** additionally guarantees: for every node, all values in the left subtree are smaller, and all values in the right subtree are larger.

### Problem 7.1 — Implement a BST: `insert` and `contains`

```python
class TreeNode:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

class BST:
    def __init__(self):
        self.root = None

    def insert(self, value):
        if self.root is None:
            self.root = TreeNode(value)
            return
        current = self.root
        while True:
            if value < current.value:
                if current.left is None:
                    current.left = TreeNode(value)
                    return
                current = current.left
            else:
                if current.right is None:
                    current.right = TreeNode(value)
                    return
                current = current.right

    def contains(self, value):
        current = self.root
        while current:
            if value == current.value:
                return True
            elif value < current.value:
                current = current.left
            else:
                current = current.right
        return False
```

```python
# Test cell
bst = BST()
for v in [5, 3, 8, 1, 4, 7, 9]:
    bst.insert(v)

assert bst.contains(4) == True
assert bst.contains(9) == True
assert bst.contains(100) == False
assert bst.root.value == 5
assert bst.root.left.value == 3
assert bst.root.right.value == 8
print("All tests passed!")
```

### Problem 7.2 — In-order Traversal
Write `inorder_traversal(root)` that returns a list of values via **in-order** traversal (left → node → right). For a BST, this should return values in **sorted order** — a handy way to sanity-check your BST.

```python
def inorder_traversal(root):
    if root is None:
        return []
    return inorder_traversal(root.left) + [root.value] + inorder_traversal(root.right)
```

### Problem 7.3 — Height of a Binary Tree
Write `tree_height(root)` that returns the height of the tree (number of edges on the longest path from root to a leaf). An empty tree has height `-1`; a single-node tree has height `0`.

```python
def tree_height(root):
    if root is None:
        return -1
    return 1 + max(tree_height(root.left), tree_height(root.right))
```

### Problem 8.1 — Linear Search

```python
def linear_search(arr, target):
    for i, val in enumerate(arr):
        if val == target:
            return i
    return -1
```

### Problem 8.2 — Binary Search
Implement it **iteratively** on a sorted array. Return the index of `target`, or `-1` if not found.

```python
def binary_search(arr, target):
    low, high = 0, len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1
```

```python
# Test cell
sorted_arr = [1,3,5,7,9,11,13,15]
assert binary_search(sorted_arr, 7) == 3
assert binary_search(sorted_arr, 1) == 0
assert binary_search(sorted_arr, 15) == 7
assert binary_search(sorted_arr, 4) == -1
assert binary_search([], 5) == -1
print("All tests passed!")
```

### Reflection Q8.3
Why can't binary search be used directly on a **linked list** the same way it's used on an array, even if the linked list is sorted? (Answer in a markdown cell.)

Binary search requires **O(1) random access** — jumping directly to the middle element. A linked list has no indexing; reaching the middle node always takes O(n) pointer hops. Even on a sorted linked list, "halving" the search range costs O(n) each time, so binary search gives no speedup over linear search (both end up O(n log n) or worse). Arrays guarantee contiguous memory layout, making index access O(1) and binary search genuinely O(log n).

### Problem 9.1 — Bubble Sort

```python
def bubble_sort(arr):
    arr = arr[:]
    n = len(arr)
    for i in range(n):
        swapped = False
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swapped = True
        if not swapped:
            break
    return arr
```

```python
# Test cell
assert bubble_sort([5,2,9,1,5,6]) == [1,2,5,5,6,9]
assert bubble_sort([]) == []
assert bubble_sort([1]) == [1]
assert bubble_sort([3,2,1]) == [1,2,3]
print("All tests passed!")
```

### Problem 9.2 — Selection Sort

```python
def selection_sort(arr):
    arr = arr[:]
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr
```

```python
# Test cell
assert selection_sort([5,2,9,1,5,6]) == [1,2,5,5,6,9]
assert selection_sort([]) == []
assert selection_sort([3,2,1]) == [1,2,3]
print("All tests passed!")
```

### Problem 9.3 — Insertion Sort

```python
def insertion_sort(arr):
    arr = arr[:]
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return arr
```

```python
# Test cell
assert insertion_sort([5,2,9,1,5,6]) == [1,2,5,5,6,9]
assert insertion_sort([]) == []
assert insertion_sort([3,2,1]) == [1,2,3]
print("All tests passed!")
```

### Problem 9.4 — Merge Sort
Implement using the classic divide-and-conquer approach: split in half, recursively sort each half, then merge.

```python
def merge_sort(arr):
    if len(arr) <= 1:
        return arr
    mid = len(arr) // 2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])
    return merge(left, right)

def merge(left, right):
    result = []
    i = j = 0
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    result.extend(left[i:])
    result.extend(right[j:])
    return result
```

```python
# Test cell
assert merge_sort([5,2,9,1,5,6]) == [1,2,5,5,6,9]
assert merge_sort([]) == []
assert merge_sort([1]) == [1]
assert merge_sort([3,2,1]) == [1,2,3]
print("All tests passed!")
```

### Problem 9.5 — Quick Sort
Implement using a pivot (e.g. last element or middle element) and partitioning.

```python
def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quick_sort(left) + middle + quick_sort(right)
```

```python
# Test cell
assert quick_sort([5,2,9,1,5,6]) == [1,2,5,5,6,9]
assert quick_sort([]) == []
assert quick_sort([1]) == [1]
assert quick_sort([3,2,1]) == [1,2,3]
print("All tests passed!")
```

### Reflection Q9.6
Merge sort always runs in O(n log n), but quick sort can degrade to O(n²) in the worst case. What input would cause quick sort's worst case if you always pick the **last element** as the pivot? (Answer in markdown.)

If you always pick the **last element** as the pivot, an already-sorted (or reverse-sorted) array causes worst-case O(n²) behavior. For example, `[1, 2, 3, 4, 5]`: the pivot is `5`, the "less than" partition has all 4 remaining elements, and the "greater than" partition is empty. Every partition step only removes one element, so you get n recursive calls each doing O(n) work — n × n = O(n²). A reverse-sorted array `[5, 4, 3, 2, 1]` with last-element pivot triggers the same pattern.

### Problem 10.1 — Factorial

```python
def factorial(n):
    if n <= 1:
        return 1
    return n * factorial(n - 1)
```

### Problem 10.2 — Fibonacci (with memoization)
Write `fibonacci(n)` returning the `n`-th Fibonacci number (`fib(0)=0, fib(1)=1`). Use a dictionary to **memoize** results so it runs in O(n) instead of the naive O(2^n).

```python
def fibonacci(n, memo=None):
    if memo is None:
        memo = {}
    if n in memo:
        return memo[n]
    if n == 0:
        return 0
    if n == 1:
        return 1
    memo[n] = fibonacci(n - 1, memo) + fibonacci(n - 2, memo)
    return memo[n]
```

```python
# Test cell
assert fibonacci(0) == 0
assert fibonacci(1) == 1
assert fibonacci(10) == 55
assert fibonacci(30) == 832040  # should be fast thanks to memoization
print("All tests passed!")
```

### Problem 10.3 — Sum of Digits (recursive)
Write `sum_of_digits(n)` that recursively sums the digits of a non-negative integer.

Example: `sum_of_digits(12345)` → `15`

```python
def sum_of_digits(n):
    if n < 10:
        return n
    return (n % 10) + sum_of_digits(n // 10)
```

```python
# Test cell
assert sum_of_digits(12345) == 15
assert sum_of_digits(0) == 0
assert sum_of_digits(9) == 9
assert sum_of_digits(100) == 1
print("All tests passed!")
```

### Problem 10.4 — Generate All Subsets (Power Set)
Write `subsets(nums)` that recursively returns all possible subsets of a list (the power set), including the empty subset and the full list. Order within/between subsets doesn't matter (tests normalize before comparing).

Example: `subsets([1,2])` → `[[], [1], [2], [1,2]]`

```python
def subsets(nums):
    if not nums:
        return [[]]
    rest = subsets(nums[1:])
    return rest + [[nums[0]] + s for s in rest]
```

```python
# Test cell
def normalize_subsets(subs):
    return sorted(sorted(s) for s in subs)

result = subsets([1,2])
expected = [[], [1], [2], [1,2]]
assert normalize_subsets(result) == normalize_subsets(expected)

result3 = subsets([1,2,3])
assert len(result3) == 8  # 2^3 subsets
print("All tests passed!")
```
